Loading the Pre trained model

```python
from transformers import MarianMTModel, MarianTokenizer
model_name = 'Helsinki-NLP/opus-mt-en-es'

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
text = input("Enter any text in the English language:")
translated = tokenizer.encode (text , return_tensors= 'pt')

translated_text = model.generate(translated , max_length= 40)

output = tokenizer.decode(translated_text[0], skip_special_tokens=True)
print(output)
```
